```python
df['type'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title("Count of Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter

all_genres = ",".join(df['listed_in'].dropna()).split(", ")
top_genres = Counter(all_genres).most_common(10)

genres, counts = zip(*top_genres)
plt.barh(genres, counts, color='green')
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genres")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Handle missing values in 'date_added' before conversion
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')  # 'coerce' will replace errors with NaT

# Extract year safely
df['year_added'] = df['date_added'].dt.year

# Drop rows where 'year_added' is NaN
df = df.dropna(subset=['year_added'])
df['year_added'] = df['year_added'].astype(int)  # Convert to integer

# Count movies and TV shows per year
yearly_counts = df['year_added'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-')
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.title("Number of Netflix Titles Added Per Year")
plt.grid(True)

# Save the graph
plt.savefig("images/netflix_trends.png")

# Show the graph
plt.show()

```
