```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("netflix_titles.csv")  

# View first few rows
df.head()
```

```python
# Check the number of rows and columns
df.shape
```

```python
# Display column names
df.columns
```

```python
# Show general info about the dataset
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# OR Fill missing values with a placeholder
df.fillna("Unknown", inplace=True)

# Verify missing values again
df.isnull().sum()
```

```python
df['type'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title("Count of Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter

all_genres = ",".join(df['listed_in'].dropna()).split(", ")
top_genres = Counter(all_genres).most_common(10)

genres, counts = zip(*top_genres)
plt.barh(genres, counts, color='green')
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genres")
plt.show()
```
